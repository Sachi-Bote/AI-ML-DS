```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Practice NumPy

#Data Structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 180000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Birger King
    [5, 160000, 185000, 205000, 230000]  #Dominos
])

print("Sales Data Analysis:\n", sales_data)
print("Shape of Data:", sales_data.shape)

#Print First 3 Restaurants Data
print("First 3 Restaurants Data:\n", sales_data[:3])
```

```python
#Total sales per year
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
print("Total Sales Per Year (2021-2024):", total_sales_per_year)

#4 years total sales per restaurant
total_sales_per_restaurant = np.sum(sales_data[:, 1:], axis=1)
print("Total Sales Per Restaurant (2021-2024):", total_sales_per_restaurant)

#Minimum and maximum sales per restaurant
min_sales_per_restaurant = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales Per Restaurant (2021-2024):", min_sales_per_restaurant)

max_sales_per_restaurant = np.max(sales_data[:, 1:], axis=1)
print("Maximum Sales Per Restaurant (2021-2024):", max_sales_per_restaurant)

#Maximum sales per year
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales Per Year (2021-2024):", max_sales_per_year)

#Average sales per restaurant
average_sales_per_restaurant = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales Per Restaurant (2021-2024):", average_sales_per_restaurant)

#Average sales per year
average_sales_per_year = np.mean(sales_data[:, 1:], axis=0)
print("Average Sales Per Year (2021-2024):", average_sales_per_year)
```

```python
#Cumulative sales per restaurant
cumulative_sales_per_restaurant = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales Per Restaurant (2021-2024):\n", cumulative_sales_per_restaurant)

#Plotting cumulative sales for each restaurant
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales_per_restaurant, axis=0), marker='o')
plt.title('Average Cumulative Sales Per Year (2021-2024)')
plt.xlabel('Years (2021-2024)')
plt.ylabel('Average Cumulative Sales')
```

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])   

print("Vector Addition:", vector1 + vector2)
print("Vector Subtraction:", vector2 - vector1)
print("Vector Multiplication:", vector1 * vector2)
print("Vector Division:", vector2 / vector1)
print("Dot Product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between Vectors (radians):", angle)
```
