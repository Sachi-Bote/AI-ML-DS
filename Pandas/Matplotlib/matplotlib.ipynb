```python
import pandas as pd
import  numpy as np 
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,6]

plt.plot(x, y, label='2x', color='red', linestyle='--', marker='o', markersize=5, markerfacecolor='blue')

plt.title("First Plot", fontdict={'fontsize': 20, 'fontweight': 'bold', 'fontname': 'Comic Sans MS'})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.show()
```
